### <b><span style='color:darkblue'>1. What factors are currently driving or hindering customer loyalty at Company XYZ?</span></b>

To determine the factors driving or hindering customer loyalty at Company XYZ, we can analyze several key aspects of the dataset and customer behavior:

```python
# Plotting the purchase frequency
plt.figure(figsize=(10, 6))
df['Frequency of Purchases'].value_counts().plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Customer Purchase Frequency')
plt.xlabel('Frequency')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The bar chart reveals that most customers at Company XYZ make purchases less frequently, with a significant portion buying every three months or annually. This suggests that the company may have a more transactional relationship with its customers, rather than a deeply loyal customer base that makes frequent purchases.

```python
# Plotting subscription status
plt.figure(figsize=(8, 8))
df['Subscription Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Subscription Status of Customers')
plt.ylabel('')
plt.show()
```

The pie chart indicates that a large percentage of customers are not subscribed to any loyalty programs or services offered by Company XYZ. This highlights a potential gap in customer engagement, where the existing subscription offerings may not be sufficiently appealing or effectively marketed.

```python
# Plotting review ratings
plt.figure(figsize=(10, 6))
plt.hist(df['Review Rating'], bins=10, color='lightblue', edgecolor='black')
plt.title('Distribution of Review Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Reviews')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The histogram shows that the majority of customer reviews are positive, with most ratings clustering around 4 or 5 stars. This reflects general customer satisfaction with the products and services provided by Company XYZ, which is a strong indicator of customer loyalty. However, the presence of lower ratings suggests that there are areas where improvements are needed, such as product quality or customer service.

```python
# Plotting the use of discounts and promo codes
plt.figure(figsize=(10, 6))
df[['Discount Applied', 'Promo Code Used']].apply(pd.Series.value_counts).plot(kind='bar', color=['lightblue', 'lightgreen'], edgecolor='black')
plt.title('Discounts and Promo Code Usage')
plt.xlabel('Usage')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The bar chart shows that discounts and promo codes are used by a substantial number of customers, indicating that these offers are an important driver of purchases. This suggests that many customers are price-sensitive and respond well to promotional incentives.

**Conclusion**

The visualizations highlight key factors influencing customer loyalty at Company XYZ. The infrequent purchase patterns and low subscription rates suggest areas where the company could improve customer engagement. At the same time, the generally positive review ratings reflect customer satisfaction, while the effective use of discounts indicates a strong response to promotional efforts. To build deeper customer loyalty, Company XYZ should focus on enhancing subscription offerings, addressing areas of dissatisfaction in product quality or service, and optimizing promotional strategies to encourage more frequent purchases.

### <b><span style='color:darkblue'>2. What role do discounts and promotional offers play in driving repeat purchases?</span></b>

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Discount Applied', y='Previous Purchases', data=df, palette='viridis', edgecolor='black')
plt.title('Impact of Discounts on Repeat Purchases')
plt.xlabel('Discount Applied')
plt.ylabel('Average Number of Previous Purchases')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight**

The bar chart shows the relationship between the use of discounts and the average number of previous purchases. It indicates that customers who use discounts tend to have a higher number of previous purchases, suggesting that discounts and promotions are effective in driving repeat purchases.

<a id="basic"></a>
### <b><span style='color:darkblue'>3. Which customer demographics are most likely to become repeat buyers, and how can this information be used to tailor loyalty programs?</span></b>

```python
age_groups = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65, 75], labels=['18-25', '26-35', '36-45', '46-55', '56-65', '66-75'])
plt.figure(figsize=(10, 6))
sns.barplot(x=age_groups, y='Previous Purchases', data=df, palette='coolwarm', edgecolor='black')
plt.title('Repeat Purchases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Number of Previous Purchases')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight**

The bar chart illustrates the average number of previous purchases across different age groups. It reveals that certain age groups, particularly those in the 46-55 range, tend to have higher repeat purchase rates. This suggests that middle-aged customers may exhibit higher loyalty, making them an important demographic for targeted loyalty programs.

<a id="basic"></a>
### <b><span style='color:darkblue'>4. How does customer satisfaction with shipping and delivery influence repeat purchase behavior?</span></b>

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Shipping Type', y='Previous Purchases', data=df, palette='Set2')
plt.title('Shipping Type vs. Repeat Purchases')
plt.xlabel('Shipping Type')
plt.ylabel('Number of Previous Purchases')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight** 

The box plot shows the relationship between different shipping types and the number of previous purchases. Customers who choose certain shipping types, such as "Next Day Air" or "Express," may have slightly higher repeat purchase rates, which could indicate satisfaction with faster delivery options.
